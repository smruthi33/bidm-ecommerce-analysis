## Import necessary packages

```python
import pandas as pd
```

```python
import numpy as np
```

## Import the datasets

```python
products = pd.read_csv("./products.csv")
products.head()
```

```python
order_prior=pd.read_csv("./order_products__prior.csv")
order_prior.head()
```

```python
aisle=pd.read_csv("./aisles.csv")
aisle
```

## Merge the Datasets

```python
merge=pd.merge(products, order_prior, on="product_id")
merge.head()
```

```python
merge=merge.groupby("aisle_id").size().reset_index(name="aisle_count")
merge
```

```python
merge=merge.merge(aisle[["aisle_id","aisle"]], on="aisle_id")
merge
```

```python
bins = [0, merge["aisle_count"].quantile(0.25), merge["aisle_count"].quantile(0.75), max(merge["aisle_count"])]
names = ["low","medium","high"]

merge['Aisle_consumption'] = pd.cut(merge['aisle_count'], bins, labels=names)
merge


```

```python
merge.to_csv("./aisle_consumption.csv", index=False)
```
